```python
from mxnet import autograd,gluon,init,nd
from mxnet.gluon import nn,data as gdata,loss as gloss
import numpy as np
import time
from IPython import display
from matplotlib import pyplot as plt
```

# 1. 读取数据集
---
使用数据集是 来自NASA的 测试不同飞机机翼噪声的数据

```python
# genfromtxt中的delimiter意思是，用于分隔值的字符串是什么。
data = np.genfromtxt("D:/plantainz/OuO_DeepLearning/DataSet/airfoil_self_noise.dat",delimiter='\t')
data = (data - data.mean(axis=0)) / data.std(axis=0)

features = nd.array(data[:1500,:-1])
labels = nd.array(data[:1500,-1])

features.shape
print(features)
```

# 2. sgd设置 
---

```python
# states：状态输入
# hyperparams：超参数字典
# params:[w,b]
def sgd(params,states,hyperparams):
    for p in params:
        p[:] -= hyperparams['lr'] * p.grad
```

# 3. 手动实现原始训练函数
---

```python
def linreg(X,w,b):
    return nd.dot(X,w) + b
```

```python
def squared_loss(y_hat,y):
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

```python
def set_figsize(figsize = (3.5,2.5)):
    display.set_matplotlib_formats('svg') # 用矢量图显示！！
    plt.rcParams['figure.figsize'] = figsize
```

```python
def train(trainer_fn, states, hyperparams,
          features, labels,
          batch_size=10, num_epochs=2):
    # 初始化模型
    net = linreg
    loss = squared_loss

    # 网络参数
    w = nd.random.normal(scale=0.01, shape=(features.shape[1], 1))
    b = nd.zeros(1)
    w.attach_grad()
    b.attach_grad()

    def eval_loss():
        return loss(net(features, w, b), labels).mean().asscalar()
    
    ls = [eval_loss()]
    print('ls = [eval_loss()],origin:',ls)
    data_iter = gdata.DataLoader(gdata.ArrayDataset(features,labels),
                                 batch_size,shuffle=True)
    
    for _ in range(num_epochs):
        start = time.time()
        for batch_i,(X,y) in enumerate(data_iter):
            with autograd.record():
                l = loss(net(X,w,b),y).mean() # 使用平均损失
            l.backward()
            trainer_fn([w,b],states,hyperparams)
            
            if(batch_i+1) * batch_size % 100 == 0: # 每一百个样本记录当下训练误差
                ls.append(eval_loss())
    # 打印结果
    print('plot===> 0,num_epochs:',num_epochs)
    print('len of ls:',len(ls))
    print('ls = [eval_loss()],after training:',ls)
    print('\n loss: %f,%f sec per epoch' %(ls[-1] , time.time()-start))
    
    set_figsize()
    plt.plot(np.linspace(0,num_epochs,len(ls)) , ls) # 这个plot，没懂第三个什么意思。
    plt.xlabel('epoch')
    plt.ylabel('loss')
    
```

# 4. 载入组件生成训练函数
---

```python
def train_format(trainer_fn,states,hyperparams,
         features,labels,
         batch_size=10,num_epochs=2):
    # 首先生成网络（也就是一个线性回归啦。dense(1)！！）
    net = nn.Sequential()
    net.add(nn.Dense(1))
    net.initialize(init.Normal(sigma=0.01))
    
    # 然后是损失函数！
    loss = gloss.L2Loss()
    def eval_loss():
        return loss(net(features),labels).mean().asscalar()
    ls = [eval_loss()]
    
    data_iter = gdata.DataLoader(gdata.ArrayDataset(features,labels),
                                batch_size,shuffle=True)
    trainer = gluon.Trainer(net.collect_params(),trainer_fn,hyperparams)
    
    # 设置好之后，开始epochs
    for _ in range(num_epochs):
        start = time.time()
        for batch_i , (X,y) in enumerate(data_iter): # 看好这个batch_i的操作
            with autograd.record():
                l = loss(net(X),y)
            l.backward()
            trainer.step(batch_size) # 在trainer里做梯度平均
            if (batch_i +1) * batch_size % 100 == 0:
                ls.append(eval_loss())
                
    print('after train,ls:',ls)
    print('\n loss:%f,%f sec per epoch' %(ls[-1] , time.time()-start))
    
    set_figsize()
    plt.plot(np.linspace(0,num_epochs,len(ls)),ls)
    plt.xlabel('epoch')
    plt.ylabel('loss')
```

# 5. 不同超参数下的训练结果
---
暂时使用原始训练函数！
两种训练函数其实效果一样的啦。

```python
# this is original gradient desent
# 走你！ lr = 1,batch_size = 1500
train(trainer_fn=sgd, states=None, hyperparams={'lr':1},
      features=features, labels=labels,
      batch_size=1500, num_epochs=6)
```

```python
# this is SGD!!
# 如果 lr = 0.05,batch_size = 1
train_format(trainer_fn='sgd', states=None, hyperparams={'learning_rate':0.05},
      features=features, labels=labels,
      batch_size=1, num_epochs=6)
# 多多多！流批！QuQ
```

```python
# this is mini-batch gradient decent
# 如果 lr = 0.05,batch_size = 10
train(trainer_fn=sgd, states=None, hyperparams={'lr':0.05},
      features=features, labels=labels,
      batch_size=10, num_epochs=6)
```
