```python
def set_figsize(figsize=(3.5,2.5)):
    '''set matplotlib figure size'''
    display.set_matplotlib_formats('svg')
    plt.rcParams['figure.figsize'] = figsize
```

```python
set_figsize()
img = image.imread('.\\lxg_logo.jpg')
plt.imshow(img.asnumpy()) # QuQ凌雪永远是我的快乐源泉！
```

```python
def show_images(imgs,num_rows,num_cols,scale = 2):
    figsize = (num_cols *scale, num_rows * scale)
    # 因为一张图中要画两个坐标系，所以用的是subplot[s] 而不是 subplot
    # 只是为了消除图边界的空白
    _,axes = plt.subplots(num_rows,num_cols,figsize = figsize) # _为返回的子图，ax为返回的坐标系（为一个数组） 
    print("axes:",axes)
    
    for i in range(num_rows):
        for j in range(num_cols):
            axes[i][j].imshow(imgs[i*num_cols+j].asnumpy()) 
            axes[i][j].axes.get_xaxis().set_visible(True) # 这里只是决定坐标系是否显示而已
            axes[i][j].axes.get_yaxis().set_visible(False)
    return axes 
```

```python
def apply(img,aug,num_rows = 2,num_cols = 4,scale=1.5):
    Y = [aug(img) for _ in range(num_rows*num_cols)]
    show_images(Y,num_rows,num_cols,scale)
```

```python
# 左右翻转
apply(img,gdata.vision.transforms.RandomFlipLeftRight())
```

```python
# 上下翻转
apply(img,gdata.vision.transforms.RandomFlipTopBottom())
```

```python
shape_aug = gdata.vision.transforms.RandomResizedCrop(
    # 裁剪后的图像大小为(200,200)，随机选取原图10%~100%的区域，宽高比为0.5~2
    (200, 200), scale=(0.1, 1), ratio=(0.5, 2))
apply(img, shape_aug)
```

```python
# 颜色变化
apply(img, gdata.vision.transforms.RandomBrightness(0.5))
```

```python
# 调整色调
apply(img, gdata.vision.transforms.RandomHue(0.5))
```

```python
color_aug = gdata.vision.transforms.RandomColorJitter(
    # brightness亮度，contrast对比度,saturation饱和度，hue色调
    brightness=0.5, contrast=0.5, saturation=0.5, hue=0.5)
apply(img, color_aug)
```

```python
# 但是呢，经常都是很多个增广方法叠加起来一起用
augs = gdata.vision.transforms.Compose([
    gdata.vision.transforms.RandomFlipLeftRight(), color_aug, shape_aug])
apply(img, augs)
```

```python
def load_cifar10(is_train,augs,batch_size):
    return gdata.DataLoader(
    gdata.vision.CIFAR10(train=is_train).transform_first(augs),
    batch_size=batch_size,shuffle=is_train,num_workers=num_workers)
```

### 使用多GPU训练模型
---

```python
def try_all_gpu():
    ctxes=[]
    try:
        for i in range(16): # 假设一台机器上的GPU数不超过16
            ctx = mx.gpu(i)
            # 这只是一个将数组放上gpu上的小测试。如果出错了就走下方except
            _ = nd.array([0],ctx = ctx) 
            ctxes.append(ctx)
    except mx.base.MXNetError: pass     # 如果出错了，不做任何操作
    if not ctxes: ctxes = [mx.cpu()]    # 如果没有cpu，就
    return ctxes
```

```python
# 小批量数据样本batch划分并复制到ctx变量所指定的各个显存上
def _get_batch(batch,ctx):
    features,labels = batch
    if labels.dtype!=features.dtype:
        labels = labels.astype(features.dtype)
    return (gutils.split_and_load(features,ctx),
            gutils.split_and_load(labels,ctx),features.shape[0])
```
